Q1. Data preprocessing (handling missing values, outliers, feature scaling).

a. handling missing values


```python
numeric_df = df.select_dtypes(include=[ 'float64', 'int64' ])
print("Missing values in numeric columns:\n", numeric_df.isnull().sum())
numeric_df_dropped = numeric_df.dropna()
numeric_df_filled= numeric_df.fillna(numeric_df.mean())
```

b. handling outliers

```python
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))
numeric_df_no_outliers =numeric_df [~outliers.any ( axis=1) ]
print("Data after removing outliers: \n", numeric_df_no_outliers)
```

c. feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numeric_df_scaled =pd.DataFrame(scaler.fit_transform(numeric_df_no_outliers), columns=numeric_df_no_outliers.columns)
print("Data after Min-Max scaling: Xn", numeric_df_scaled)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
numeric_df_standardized = pd.DataFrame(scaler.fit_transform(numeric_df_no_outliers), columns=numeric_df_no_outliers.columns)
print("Data after Standardization: Xn", numeric_df_standardized)
```

Q3. Building a simple linear regression model and evaluating model performance using accuracy, precision, recall, F1-score.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

df = pd.get_dummies(df, drop_first=True)
df.fillna(value=0, inplace=True)

X = df.drop(columns=['churn'])
y = df['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, y_train)

y_pred = linear_reg_model.predict(X_test)
y_pred_binary = (y_pred > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred_binary)
precision = precision_score(y_test, y_pred_binary)
recall = recall_score(y_test, y_pred_binary)
f1 = f1_score(y_test, y_pred_binary)
conf_matrix = confusion_matrix(y_test, y_pred_binary)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
print("Confusion Matrix:\n", conf_matrix)
```
