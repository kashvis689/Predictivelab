```python
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv(r'C:\Users\AxD\Downloads\bodyPerformance.csv')

# Exclude non-numeric columns for statistical analysis
numeric_data = data.select_dtypes(include=[np.number])

# 1.1 Measure of Central Tendency
mean_vals = numeric_data.mean()
geometric_mean_vals = np.exp(np.mean(np.log(numeric_data[numeric_data > 0]), axis=0))  # Geometric mean for positive values
harmonic_mean_vals = len(numeric_data) / np.sum(1.0 / numeric_data[numeric_data > 0], axis=0)  # Harmonic mean for positive values
median_vals = numeric_data.median()
mode_vals = numeric_data.mode().iloc[0]

# 1.2 Measure of Dispersion
variance_vals = numeric_data.var()
std_dev_vals = numeric_data.std()
range_vals = numeric_data.max() - numeric_data.min()
iqr_vals = numeric_data.quantile(0.75) - numeric_data.quantile(0.25)

# Calculate MAD manually
mad_vals = numeric_data.apply(lambda x: np.mean(np.abs(x - x.mean())))

skew_vals = numeric_data.skew()

# 1.3 Correlation between features
correlation_matrix = numeric_data.corr()

# Print results
print("Mean:\n", mean_vals)
print("Geometric Mean:\n", geometric_mean_vals)
print("Harmonic Mean:\n", harmonic_mean_vals)
print("Median:\n", median_vals)
print("Mode:\n", mode_vals)
print("Variance:\n", variance_vals)
print("Standard Deviation:\n", std_dev_vals)
print("Range:\n", range_vals)
print("Interquartile Range (IQR):\n", iqr_vals)
print("Mean Absolute Deviation (MAD):\n", mad_vals)
print("Skewness:\n", skew_vals)
print("Correlation Matrix:\n", correlation_matrix)
```
